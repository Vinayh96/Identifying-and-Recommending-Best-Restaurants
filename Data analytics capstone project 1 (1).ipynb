```python
##Pie chart to show percentage of restaurants which allow table booking and those which don't
labels = ['table_booking','no_table_booking']
value = [table_booking,no_table_booking]
plt.figure(figsize = [6,6])
plt.pie(value,labels = labels,startangle = 90,autopct = '%2.1f%%',explode = [0.1,0],shadow = True)
plt.title('table_booking vs no_table_booking')
plt.show()
```

Above pie chart clearly shows that only 12.13% restaurants allow table booking and majority of 87.87% don't allow.

## Calculate the difference in number of votes for the restaurants that deliver and the restaurants that do not deliver

```python
deliver_booking=rest_data[rest_data['Has_Online_delivery']=='Yes']['Votes'].sum()
deliver_booking
```

```python
no_deliver_booking=rest_data[rest_data['Has_Online_delivery']=='No']['Votes'].sum()
no_deliver_booking
```

```python
print(abs(deliver_booking-no_deliver_booking))
```

```python
labels = ['Has_Online_delivery','no_Online_delivery']
value = [deliver_booking,no_deliver_booking]
plt.figure(figsize = [6,6])
plt.pie(value,labels = labels,startangle = 90,autopct = '%2.1f%%',explode = [0.1,0],shadow = True)
plt.title('number of votes for the restaurants that deliver and the restaurants that do not deliver')
plt.show()
```

It means that 65.4% votes have been given to those restaurants who don't have online delivery and those with online delivery have received voting of 34.6%.
It means that restaurants with no online delivery are getting more votes.
